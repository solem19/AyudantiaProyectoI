```python
f=lambda x: np.exp(-((x[0]*x[1])**2+x[0]**2+x[1]**2-8*x[0]-8*x[1])/2) # Esta función de densidad conjunta es la vista en clase, pero chequen que no tiene ninguna constante c. 
```

```python
#La función simuladora, donde N es la cantidad de variables que va a generar, contando el quemado;el valor inicial que lo
#pueden modificar, el periodo de quemado que tiene que ser menor que N porque si no truena, y la función de densidad a simular. 

#Chequen cómo lo que está adentro del for es el algoritmo que vimos en clase; el algoritmo sólo va a devolver las variables
#aleatorias que salgan DESPUÉS del periodo de quemado. 
def genf2d(N,a=1,valorinicial=np.array([0,0]),burn=1000, f=lambda x: np.exp(-((x[0]*x[1])**2+x[0]**2+x[1]**2-8*x[0]-8*x[1])/2) ):
    Variable=[]
    X=valorinicial
    for i in range(0,N):
        Z=a*np.random.normal(size=2)
        C=X+Z
        print([c for c in C])
        alpha=min( (f(C)/f(X)), 1  )
        U=np.random.uniform(size=1)[0]
        if U<alpha:
            X=C
        if i>=burn:
            Variable.append(X)
    return Variable
        
    
    
```

```python
#Un ejemplo para simular 1000 variables aleatorias de f: 
genf2d(2000)
```
